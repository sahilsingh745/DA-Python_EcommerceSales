```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To show all the Columns
pd.set_option("display.max_columns",24)
```

```python
dataset = pd.read_excel("Superstore_USA.xlsx")
```

```python
dataset.head(5)
```

```python
#no of columns and rowa
dataset.shape
```

```python
#For finding missing value - analysis_will return total of null values of each columns ****
dataset.isnull().sum()
```

```python
# Filling blank values with mean of the column and saving it permanently
dataset["Product Base Margin"].fillna(dataset["Product Base Margin"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

# ORDER PRIORITY

```python
# analyssing order priority - Giving Total count of each value.
dataset["Order Priority"].value_counts()
```

```python
#speciphying the types of the data
dataset["Order Priority"].unique()
```

```python
#now repacing # 'Critical ' with  'Critical'
dataset["Order Priority"] = dataset["Order Priority"].replace('Critical ','Critical')
```

```python
#to dcrease the size of thee graph
plt.figure(figsize=(5,4))
#seaborn libraru used for pltting of col: Order Priority and fron given dataset
sns.countplot(x = "Order Priority",data = dataset)
#For tilte of graph
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")
#to show the graph
plt.show
```

# Shipping Mode

```python
#counting the shiping modes
dataset["Ship Mode"].value_counts()
```

```python
#Defining the axes x and y respectively
x = dataset["Ship Mode"].value_counts().index
y = dataset["Ship Mode"].value_counts().values
```

```python
#using a pie chart
plt.figure(figsize=(5,4))
plt.pie(y,labels = x,startangle = 60,autopct =" %0.2f%%")  # passing 'y' as making pie in the basis of values of 'indexes' and defining labels
plt.legend(loc = 2)
plt.show
```

```python
#SHIPPING MODE on the bsis of PRODUCT CATEGORY
plt.figure(figsize=(5,4))
sns.countplot(x="Ship Mode",data=dataset,hue = "Product Category") #hue: paameter helps to differentiate more on the basis of other parameter
plt.show
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Customer Segment",data=dataset)
plt.show
```

# Product Category

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Product Category",data=dataset)
plt.show
```

```python
# Now analysing SUB produc category only for ::: OFFICE SUPPLIES
plt.figure(figsize=(6,10))
sns.countplot(x="Product Category",data=dataset[dataset["Product Category"] == "Office Supplies"],hue = "Product Sub-Category")
                                            #Bounding our data set just for "PRODUCT CATEGOR'S" "OFFICE SUUPLES"
plt.show
```

# Annual Sales

```python
#Changing/Replacing  the whole column 
dataset["Order Year"] = dataset["Order Date"].dt.year
dataset.shape
```

```python
dataset["Order Year"].value_counts()  #Year be year order vale is increasing
```
